```python
##Lop controls
##Break and continue
fruits_list=["Mango","Cherry","Apple","Papaya","Banana"]
```

```python
## break terminate the loop
for i in fruits_list:
    if i=="Apple":
        print("The fruit is apple")
        break
    print(i)
```

```python
for i in fruits_list:
    if i=="Apple":
        print("The fruit is apple")
        pass
    print(i)
```

```python
##continue skip the loop
for i in fruits_list:
    if i=="Apple":
        print("The fruit is apple")
        continue
    print(i)
```
